```python
import chardet
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
```

```python
with open('opendata.csv', 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large


file = pd.read_csv('opendata.csv', encoding=result['encoding'])
df = pd.DataFrame(file)
```

```python
regions = df['region'].unique()
```

```python
names = df['name'].unique()
names
```

```python
my_region = 'Россия'
my_parametr = 'Средняя сумма заявки на потребительский кредит'
my_date1 = '2013-01-01'
my_date2 = '2020-01-01'
```

```python
X1 = df[(df['region'] == my_region) & (df['name'] == my_parametr) & ((df['date'] >= my_date1) & (df['date'] <= my_date2))]
```

```python
X = X1['date']
y = X1['value']
```

```python
plt.plot(X, y)
#plt.subplots_adjust(wspace = 1, hspace=1)
#fig = plt.figure(figsize=(18,18), dpi=100)
#fig, ax = plt.subplots(figsize=(20, 10))
```
